Let's make a more **complex calculation with lambda function**

**Recap**: A lambda function is a way of defining a function in a single line of code. Usually, we would assign them to a variable

`mylambda = lambda x: (x * 2) + 3 `

`print(mylambda(5))`


sintax lambda for IF statements:

`lambda x: [OUTCOME IF TRUE] if [CONDITIONAL] else [OUTCOME IF FALSE]`




````python
#Let's print in a new column "Low" if the rating is < 3, else "High"

df["Review_score"] = df["Review Rating"].apply(lambda rating: "Low" if rating < 3 else "High")


#Let's play around with integers, even if it doesn't make sense, let's substract the second word in the column "Location". Some locations like Rhode Island, have 2 words.

df["Location_2nd_word"] = df.Location.apply(lambda x: x.split(" ")[-1] if " " in x else "")

df.head(5)

#To make it work I had to write an if statement, cause otherwise it was picking up also those with 1 single word in Location

````

````python
#Using lambda to mofidy a table on row basis and not the entire column

#We can also operate on multiple columns at once. If we use apply without specifying a single column and add the argument axis=1, the input to our lambda function will be an entire row, not a column. To access particular values of the row, we use the syntax row.column_name or row[‘column_name’].

#Data relies on automatic input but the survey in Kentucky was glitched put 1 point less to each rating in Kentucky. Let's add 1 point to Kentucky locations.


df["New_rating"] = df.apply(lambda row:
                                  row["Review Rating"]+1
                                  if row["Location"] == "Kentucky"
                                  else row["Review Rating"],
                                  axis=1)

df.head(4)


#Let's try this same logic for the above df["Location_2nd_word"]

df["Location_2nd_word"] = df.apply(lambda row:
                                           row["Location"].split(" ")[-1]
                                           if (" ") in row["Location"]
                                            else None,
                                   axis=1)

df.head()




````

````python
#Renaming columns

#To change all column names at the same time you can use .columns

df1 = pd.DataFrame({
    'name': ['John', 'Jane', 'Sue', 'Fred'],
    'age': [23, 29, 21, 18]
})
df1.columns = ['First Name', 'Age']

#You also can rename individual columns by using the .rename method. Pass a dictionary like the one below to the columns keyword argument:
#Using rename with only the columns keyword will create a new DataFrame, leaving your original DataFrame unchanged. That’s why we also passed in the keyword argument inplace=True. Using inplace=True lets us edit the original

df1 = pd.DataFrame({
    'name': ['John', 'Jane', 'Sue', 'Fred'],
    'age': [23, 29, 21, 18]
})
df1.rename(columns={
    'name': 'First Name',
    'age': 'Age'},
    inplace=True)

#Let's do this with our table

df.rename(columns = {"Purchase Amount (USD)":"Total_USD", "Subscription Status":"Subscription_Status"}, inplace = True )

df.head(4)

````

````python
#Let's write a function to change the name of all columns where there's more than one word, and merge it with the next one with _

df.columns = df.columns.map(lambda col: col.replace(" ", "_"))

#Data cleaning: let's make sure that none of the data input in all columns has unnecessary spaces before or after

df.columns = df.columns.map(lambda col: col.strip())

df.head()
````

````python
#Marketing requested a new column that creates a statement saying "The customer of [Age] years, from [Location] purchased a [Item_Purchased]"

statement = lambda row: 'The customer of {} years, from {} purchased a {}'.format(row.Age, row.Location, row.Item_Purchased)

df["Marketing statement"]= df.apply(statement, axis=1)

df.head(4)
````

````python
#Let's calculate the median of the Purchase_Amount_(USD)

df["Purchase_Amount_(USD)"].mean()

#Let's see how many unique values there are per Item_Purchased

df["Item_Purchased"].nunique()

#Let's see the unique values in Item_Purchased

df["Item_Purchased"].unique()

#Other commands are:

#mean()	Average of all values in column
#std()	Standard deviation
#median()	Median
#max()	Maximum value in column
#min()	Minimum value in column
#count()	Number of values in column
#nunique()	Number of unique values in column
#unique()	List of unique values in column


# We can also use group by in Pandas!

# Syntax  = df.groupby('column1').column2.measurement()

# Let's calculate the average review grouping by every item purchased

df.groupby("Item_Purchased").Review_Rating.mean()

# When we use groupby Pandas creates a new Series, not a DataFrame. In order to create a Dataframe, we have to use reset_index() at the end

Review_per_Item = df.groupby("Item_Purchased").Review_Rating.mean().reset_index()

# Now we have a dataframe and we can rename the columns

Review_per_Item = Review_per_Item.rename(columns={"Item_Purchased": "Item_name", "Review_Rating": "Average_Rating"})

print(Review_per_Item)
````

````python
# Sometimes, the operation that you want to perform is more complicated than mean or count.
# In those cases, you can use the apply method and lambda functions, just like we did for individual column operations.

#Let's get the 75 percentile of the age per item purchased. So we see which age groups buy which items

import numpy as np

age_item = df.groupby('Item Purchased').Age.apply(lambda x: np.percentile(x, 75)).reset_index()

#Let's get a total sum of Purchase Amount (USD) per item purchased

total_amount_per_item = df.groupby("Item Purchased")["Purchase Amount (USD)"].apply(lambda x: sum(x)).reset_index()

#Let's do the same but limit the search only for male

total_amount_per_item_male = df[df["Gender"] == "Male"].groupby("Item Purchased")["Purchase Amount (USD)"].apply(lambda x: sum(x)).reset_index()

#We can also group by multiple columns at the same time.

#Let's get the Purchase Amound USD by color by gender.

unpivoted_total_usd_color_gender = df.groupby(["Color", "Gender"])["Purchase Amount (USD)"].apply(lambda x: sum(x)).reset_index()


#However, with several group bys, it's hard to see data properly. We would benefit from pivoting the "Gender" column, let's do it.

pivoted_total_usd_color_gender = unpivoted_total_usd_color_gender.pivot(
    columns= "Gender",
    index= "Color",
    values= "Purchase Amount (USD)"
).reset_index()

print(pivoted_total_usd_color_gender)

````
